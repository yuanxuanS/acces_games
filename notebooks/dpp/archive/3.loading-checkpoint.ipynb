# Load model from checkpoint

```python
%load_ext autoreload
%autoreload 2

import sys; sys.path.append('../../')

from hydra.utils import instantiate
import omegaconf

import torch
from tensordict.tensordict import TensorDict

from ncobench.envs import TSPEnv, DPPEnv
```

```python
# Load config

path = "/home/botu/Dev/ncobench/logs/train/runs/2023-04-14_02-30-47"

cfg = omegaconf.OmegaConf.load(path + "/.hydra/config.yaml")

# Need to make paths relative to the config file
root_dir = "../../"
cfg.env.chip_fpath = root_dir + cfg.env.chip_fpath
cfg.env.decap_fpath = root_dir + cfg.env.decap_fpath
cfg.env.freq_fpath = root_dir + cfg.env.freq_fpath

print(cfg)
```

```python
# Load checkpoint
model_checkpoint = torch.load(path + "/checkpoints/last.ckpt")

# We instantiate the policy only
model = instantiate(cfg.model.policy)
```

```python
# Clean up the state dict: we need to remove the model.policy. prefix as well as the model. prefix afterwards
state_dict = model_checkpoint['state_dict']
# replace all model.policy. keys with () 
state_dict = {k.replace('model.policy.', ''): v for k, v in state_dict.items()}
# remove all keys with model
state_dict = {k: v for k, v in state_dict.items() if 'model' not in k}

model.load_state_dict(state_dict)
```

```python
env = model.env
state = env.reset(batch_size=10)
out = model(state)
```

```python
state_idx = -1
td = state[state_idx]
decaps = out['actions'][state_idx].cpu()
probe = td['probe'].cpu()
action_mask = td['action_mask'].cpu()

env.render(decaps, probe, action_mask)
print("Reward: %.3f" % out['reward'][state_idx].item())
```

## Testing from DevFormer data

```python
# load pkl data
import numpy as np

test_data = np.load("test_data.pkl", allow_pickle=True)
test_data.shape

# We need to convert the data. First two rows are locations,
# last row is type. 0: available, 1: keepout, 2: probe, 3: decap

locations = test_data[:, :, :2]

keepouts = test_data[:, :, 2] == 1
probes = test_data[:, :, 2] == 2

action_mask = np.ones_like(test_data[:, :, 0], dtype=np.bool8)
action_mask = np.logical_and(action_mask, ~keepouts)
action_mask = np.logical_and(action_mask, ~probes)

# select probe index for the state
probes = np.where(probes)[1]

init_obs = TensorDict({
    "observation": torch.from_numpy(locations).float() / 10, # normalize
    "action_mask": torch.from_numpy(action_mask).bool(),
    "probe": torch.from_numpy(probes).long().unsqueeze(-1),
}, batch_size=[test_data.shape[0]])
```

```python
init_state = env.reset(init_obs=init_obs)
out = model(init_state, decode_type="greedy") # set to greedy decoding

print("Average reward: %.3f" % out['reward'].mean())
```

## Plot some results


```python
plot_num = 10

for i in range(plot_num):
    td = init_state[i]
    decaps = out['actions'][i].cpu()
    probe = td['probe'].cpu()
    action_mask = td['action_mask'].cpu()

    env.render(decaps, probe, action_mask)
    print("Reward: %.3f" % out['reward'][i].item())
```
